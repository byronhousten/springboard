```python
%matplotlib inline
import matplotlib.pyplot as plt
import wordcloud as wc
```

```python
text = open("job_descriptions.txt", encoding='utf-8').read()
```

```python
wordcloud = wc.WordCloud().generate(text)
```

```python
plt.imshow(wordcloud)
plt.axis("off")
```

```python
wordcloud = wc.WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
